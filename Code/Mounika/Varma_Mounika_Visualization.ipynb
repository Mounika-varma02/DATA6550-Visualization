## Titanic Dataset Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("C:/Users/surpa/Downloads/train.csv")
df.head()
```

### Accurate Visualizations

#### Survival Rate by Gender

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=df)
plt.title("Survival Rate by Gender")
plt.ylabel("Survival Rate")
plt.show()
```

This bar chart shows how many men and women who were on the Titanic lived. The y-axis starts at zero, which is the right way to show the difference between groups. The chart makes it very clear that women passengers had a much better chance of survival than men passengers. This visualization is ethical and gives a clear and honest picture of the data because it uses the full dataset and does not change the scale.

#### Survival Rate by Passenger Class

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Pclass", y="Survived", data=df)
plt.title("Survival Rate by Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

This chart shows the survival rates for first, second, and third-class passengers. It shows that people in first class had the best chance of living, while people in third class had the worst. The graph shows all classes of passengers and uses the same scale for all of them, which makes it easy to compare them. This makes the picture right, clear, and morally right.

### Misleading Visualizations

#### Manipulated Y-Axis

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=df)
plt.ylim(0.4, 0.8)
plt.title("Survival Rate by Gender (Misleading)")
plt.show()
```

This graph shows how many people of each gender survived, but the y-axis starts at a higher value instead of zero. This makes the difference between the survival rates of men and women look bigger than it really is. The data itself hasn't changed, but the way it is shown makes it hard to understand. This shows how changing the scales on the axes can confuse people.

#### Cherry-Picked Data (First Class Only)

```python
first_class = df[df["Pclass"] == 1]

plt.figure(figsize=(6,4))
sns.barplot(x="Sex", y="Survived", data=first_class)
plt.title("Survival by Gender (First Class Only)")
plt.show()
```

This chart only shows first-class passengers; it doesn't show second- or third-class passengers. It doesn't give a full picture of survival patterns because it only shows part of the dataset. This selective presentation can trick people into thinking that survival rates are higher than they really are by leaving out important information. It shows how picking and choosing data can lead to biased conclusions.

### Conclusion:
This project shows how the way you choose to visualize data can change how you understand it. Accurate visualizations make it easy to see patterns in the Titanic dataset, like how men and women survived at different rates and how different classes of passengers survived. But misleading visualizations show how making design choices like changing the scales of axes or only showing some data can make things harder to understand. Even if the data stays the same, how it is presented can have a big impact on how people see it. So, for ethical visualization to work, it needs to be clear, properly scaled, and have all the information it needs. When you design carefully, data tells the truth instead of being biased.
